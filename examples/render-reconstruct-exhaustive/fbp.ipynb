```python
import tal

data = tal.io.read_capture('/media/pleiades/vault/projects/202206-tal/code/examples/render-reconstruct-exhaustive/exhaustive-scene/20240621-193320/exhaustive-scene.hdf5')  # you'll need to generate this file using "tal render exhaustive-scene"
tal.reconstruct.compensate_laser_cos_dsqr(data)
```

```python
volume_xyz = tal.reconstruct.get_volume_project_rw(data, depths=[1.0,])
```

```python
H_1_fbp = tal.reconstruct.fbp.solve(data, wl_mean=0.25, wl_sigma=0.25, volume_xyz=volume_xyz, camera_system=tal.enums.CameraSystem.DIRECT_LIGHT)
```

```python
tal.plot.amplitude_phase(H_1_fbp, title='Reconstruction')
```

```python
with tal.resources(cpu_processes=2, downscale=2):
    H_1_fbp = tal.reconstruct.fbp.solve(data, wl_mean=0.25, wl_sigma=0.25, volume_xyz=volume_xyz, camera_system=tal.enums.CameraSystem.DIRECT_LIGHT)
```

```python
tal.plot.amplitude_phase(H_1_fbp, title='Multiple threads')
```

```python
tal.plot.amplitude_phase(H_1_fbp, title='Multiple threads')
```
